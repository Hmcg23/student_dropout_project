## Feature Engineering

```python
# 1. Approval Rate (how many enrolled classes were passed)

df["approval_rate_1st_sem"] = (
    df["Curricular units 1st sem (approved)"] /
    df["Curricular units 1st sem (enrolled)"]
)

# Replace divide-by-zero errors if any students enrolled in 0 classes
df["approval_rate_1st_sem"] = df["approval_rate_1st_sem"].fillna(0)

# 2. Failed Courses

df["failed_courses_1st_sem"] = (
    df["Curricular units 1st sem (enrolled)"] -
    df["Curricular units 1st sem (approved)"]
)

# 3. Average First Semester Grade

df["avg_first_sem_grade"] = (
    df["Curricular units 1st sem (grade)"] /
    df["Curricular units 1st sem (approved)"]
)

# Replace divide-by-zero errors if student passed 0 classes
df["avg_first_sem_grade"] = df["avg_first_sem_grade"].fillna(0)

# 4. Financial Risk Flag
# Student is financially at risk if:
# - they are a debtor OR
# - tuition is not up to date

df["financial_risk"] = (
    (df["Debtor"] == 1) |
    (df["Tuition fees up to date"] == 0)
).astype(int)

# Check new features
new_features = [
    "approval_rate_1st_sem",
    "failed_courses_1st_sem",
    "avg_first_sem_grade",
    "financial_risk"
]

print(df[new_features].head())
```

## Data Encoding and Final Cleaning

### Subtask:
Remove unnecessary columns and map the 'Target' categorical variable into numerical values.


**Reasoning**:
I need to remove the columns identified in the plan and encode the 'Target' column to make it suitable for machine learning models. I will also drop 'Is_Dropout' which was created for EDA.



```python
import pandas as pd

# List of columns to drop based on the plan and EDA notes
columns_to_drop = ['Application mode', 'Nacionality', 'Application order', 'Is_Dropout']

# Filter columns that actually exist in the dataframe to avoid errors
existing_cols_to_drop = [col for col in columns_to_drop if col in df.columns]
df_prepared = df.drop(columns=existing_cols_to_drop)

# Encode the Target variable: Dropout=0, Enrolled=1, Graduate=2
target_mapping = {'Dropout': 0, 'Enrolled': 1, 'Graduate': 2}
df_prepared['Target'] = df_prepared['Target'].map(target_mapping)

print("Columns removed:", existing_cols_to_drop)
print("Target variable distribution:\n", df_prepared['Target'].value_counts())
display(df_prepared.head())
```

## Feature Scaling

### Subtask:
Standardize all features in the prepared dataset using StandardScaler.


**Reasoning**:
I am separating the features from the target and applying the standard scaler to normalize the numerical data as instructed.



```python
from sklearn.preprocessing import StandardScaler

# Separate features and target from df_prepared
X = df_prepared.drop(columns=['Target'])
y = df_prepared['Target']

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the features
X_scaled_array = scaler.fit_transform(X)

# Create a new dataframe for scaled features
df_scaled = pd.DataFrame(X_scaled_array, columns=X.columns)

print("Standardization complete.")
print("Shape of scaled features:", df_scaled.shape)
display(df_scaled.head())
```

## K-Means Clustering for Feature Engineering

### Subtask:
Apply K-Means clustering to the standardized data and add the cluster labels as a new feature to the dataset.


**Reasoning**:
I will first implement the Elbow Method to determine the optimal number of clusters for the scaled student data.



### Feature Engineering Logic: Clustering & RFE
In this preparation phase, we are using two complementary techniques to optimize our dataset:

**K-Means Clustering (Feature Creation)**: This is an unsupervised method that groups students based on similarities across all features. By adding the Cluster label as a new feature, we provide the model with a 'synthetic' variable that captures complex, non-linear relationships between students that a single raw feature might miss.

**Recursive Feature Elimination (RFE) (Feature Selection)**: Now that we have added new features (including the cluster and semestral rates), we use RFE to identify which features actually contribute to predicting the Target. RFE works by iteratively training a model (XGBoost) and removing the least important features until the optimal subset is found.
Goal: By combining these, we ensure our model has the most powerful synthetic insights (Clustering) while maintaining a lean, high-performing feature set free of noise (RFE).

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```

**Reasoning**:
I will apply K-Means clustering with k=4 to the scaled data and append the resulting cluster labels to the `df_prepared` dataframe as a new feature.



```python
from sklearn.cluster import KMeans

# Choosing k=5 based on the elbow plot and project context
k_optimal = 5
kmeans = KMeans(n_clusters=k_optimal, random_state=42, n_init=10)
clusters = kmeans.fit_predict(df_scaled)

# Add the cluster labels to the prepared dataframe
df_prepared['Cluster'] = clusters

print(f'Added Cluster feature with {k_optimal} clusters.')
print(df_prepared['Cluster'].value_counts())
display(df_prepared.head())
```
